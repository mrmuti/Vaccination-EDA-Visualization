#### Let's get introduced to our data and make it more accurate

```python
vacc_df.info()
print("missing value:",vacc_df.isnull().values.any())
vacc_df.isnull().sum()
per_missing_data = vacc_df.isnull().sum()*100/len(vacc_df)
print("per missing data:",per_missing_data)
print("duplicate columns :",vacc_df.duplicated().any())
sns.heatmap(vacc_df.isnull())
vacc_df.dropna(axis = 0)
```
